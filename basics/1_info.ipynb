```python
# Data types 
# python provides several built-in data types
# int, str, float, bool, complex, lists, tuples, dictionaries, sets and frozensets
# Numeric types: int, float, complex
num1 = 10 # int
num2 = 12.45 # float
num3 = 45 - 6j # complex
print(num1)
print(num2)
print(num3)
# type - function returns datatype used 
print(type(num1))
print(type(num2))
print(type(num3))
```

```python
# sequence types 
# ordered collections
# list - Mutable, duplicates allowed
my_list = [1, 2, 3, 'python']
print(my_list)
print(type(my_list))
```

```python
# Operators and expressions 
# Arithematic operators 
# +, -, *, /, //, %, **
print(4+5) # 9
print(4-5) # -1
print(4*5) # 20
print(4/5) # 0.8
print(4//5) # 0
print(4 % 5) # 4
print(4 ** 5) # 1024
```

```python
# Comparion operators
# ==, >=, <=, !=, <, >
print(12>5) # True
print(12>=12) # True
print(1 == 1.0) # True
print(True != 1) # False
```

```python
# Logical operators 
# and, or, not 
a, b = True, False
print(a and b) # True 
print(a or b) # True
print(False or False) # False 
print(not True) # False
```

```python
# Assignment operators 
# =, +=, -=, *=, **=, /=, //=, %=
a = 100
a += 150
print(a)
```

```python
# bitwise operators 
print(5 & 3) # bitwise-and
print(5 | 3) # bitwise-or
print(5 ^ 3) # bitwise-xor
print(~5) # bitwise-Not
print(5 >> 1) # left shift
print(5 << 2) # right shift 
```

```python
# Identity operators
# is and is not - checks address 
a = [1,2,2,3]
b = a
c = [1,2,2,3]
print(a is b) # True
print(a is c) # False
```

```python
# Membership operators
# in, not in
print(12 in [1,2,2,3]) # False
print(1 in [1,2,3,4]) # True
```

```python
# Operator precedence 
# higher to lowest 
# () > *(expo) > +=,-=(unary) > /,//,%,* >
# +,- > <<,>> > & > ^ > | > comparsions > logical
# > assignment 
print(2 + 5 * 8) # 42
print((2 + 5) * 8) # 56 
```

```python
# Control-statements 
# if-elif-else 
# if condition:
#   # code 
# elif condition:
#   # code 
# ....
# else:
#   # code 
age = 20
if age >= 18:
    print('You are eligible to vote')
else:
    print("You are not eligible to vote")
```

```python
num = 5
if num > 0:
    print("Positive")
elif num == 0:
    print("Zero")
else:
    print("Negative")
```

```python
# while loop
# loop runs as long as condition remains true 
i = 1 
while i <= 5:
    print("Hello from iteration{}".format(i))
    i += 1
```

```python
# nested loops
for i in range(1,4):
    print(f"iteration_out_{i}")
    for j in range(1,i+1):
        print(j)
```

```python
# loop control statements
# break - terminates loop immediately
# continue - terminates current and starts next 
# pass - doesnothing 
i = 1
while i <= 10:
    if i % 2 == 0:
        i += 1
        continue
    print(i, end=' ')
    i += 1
```

```python
for i in range(1,11):
    if i == 5:
        break
    print(i,end=' ')
```
