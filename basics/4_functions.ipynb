```python
# exception handling
# try - executes instructions or catches exception
# except - handles exceptions 
# finally - executes no matter with or without exception
try:
    print(1/0)
except ZeroDivisionError as err:
    print(err)
```

```python
try:
    name = input("Enter name: ")
    if not name:
        raise ValueError("Enter valid name!")
    print(f"Hello {name}")
except ValueError as err:
    print(err)
```

```python
try:
    def myfunc():
        x = 10
        print(x)
    myfunc()
    print(x)
except NameError as err:
    print(err)
```

```python
# function
# block of statements to do specific task when called
# functions are two two types: built-in and user-defined
# user-defined functions are defined using def keyword
def sayHello():
    print("Hello")
sayHello()
```

```python
# parameters - variables defined in defining function
# default parameters - default putting value
def sayHello(name='mandy'):
    print(f"Hello {name}")
# arguments - variables passing over calling functions
# positional args - on based on position
# keyword args - on based on variable name
# *args and **kwargs - variable length of positional and keyword varibles
def total(*args):
    return sum(args)
print(total(1,2,2,3))
def info(**kwargs):
    for key, val in kwargs.items():
        print(key,'=',val)
info(name='sunny',age=22,city='hyderabad')
```

```python
# function calling itself again and again is recursive function
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n-1)
print(factorial(5))
print()

def print1ton(n):
    if n == 0:
        return
    print1ton(n-1)
    print(n)
print1ton(3)
print()

def printnto1(n):
    if n == 0:
        return
    print(n)
    printnto1(n-1)
printnto1(5)
```

```python
def total(n):
    if n == 0:
        return 0
    return n + total(n-1)
print(total(100))
```

```python
def prime(n):
    if n < 2:
        return False
    i = 2
    while i < int(n ** 0.5) + 1:
        if n % i == 0:
            return False
        i += 1
    return True

def primenumbers(a,b):
    for i in range(a,b+1):
        if prime(i):
            print(i,end=' ')

primenumbers(1,25)
```

```python
def binary_number(n: int):
    if n == 0:
        return 0
    res = ''
    while n != 0:
        res = str(n%2) + res
        n //= 2
    return int('0' + res)
# print(binary_number(0))
# print(bin(0)[2:])

for i in range(11):
    print(binary_number(i),end=' ')
```

```python
def palindrome(word):
    res = ''
    for char in word:
        res = char + res 
    return res == word
print(palindrome('amma'))
print(palindrome('racecar'))
```

```python
def factorial(n):
    fact = 1
    if n == 0:
        return fact
    for i in range(1,n+1):
        fact *= i
    return fact 
print(factorial(5))
```

```python
def fibonacci(n):
    a,b = 0, 1
    res = [a]
    for _ in range(n):
        res.append(a)
        a,b = b,a+b
    return res
print(fibonacci(5))
```

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
lst = [fibonacci(i) for i in range(6)]
print(lst)
```

```python
def armstrong(n):
    def countdigits(temp,res=0):
        while temp != 0:
            res += 1
            temp //= 10
        return res 
    def sumOfdigitsTopow(temp,digitscount,res=0):
        while temp != 0:
            digit = temp % 10
            res += (digit ** digitscount)
            temp //= 10
        return res 
    number_of_digits = countdigits(temp=n)
    summation = sumOfdigitsTopow(temp=n,digitscount=number_of_digits)
    return summation == n
print(armstrong(153))
    
```
