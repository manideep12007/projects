```python
'''Sequence of characters enclosed by quotes is called string'''
name = "python"
num = '123'
print(name)
print(num)
print(type(name))
print(type(num))
```

```python
# strings are immutable objects in python
# string methods 
str1 = 'I love programming'
# len - number of characters
length = len(str1) 
print(length)
```

```python
# strip - remove whiespaces from ends 
str1 = '  programming  '
print(str1.strip())
print(str1.lstrip())
print(str1.rstrip())
```

```python
# split - into list 
str1 = 'I learn coding in python'
lst = str1.split(' ')
print(lst)
```

```python
# upper, lower, swapcase 
str1 = 'python Programming'
print(str1.upper())
print(str1.lower())
print(str1.swapcase())
print(str1.title())
print(str1.capitalize())
```

```python
# startswith, endswith
str1 = 'Python'
print(str1.startswith('P'))
print(str1.endswith('on'))
```

```python
# replace - to replace substr or char 
str1 = 'I love coding in javascript'
print(str1.replace('javascript','python'))
```

```python
# index, find - returns index of char or substr
str1 = 'python programming'
print(str1.find('P')) # safe method
print(str1.index('p'))
print(str1.rindex('p'))
```

```python
# count - returns no of repeatations of substr or char 
str1 = 'python programming'
print(str1.count('p'))
```

```python
# join - used to form into string from list 
lst = 'Hello world'.split(' ')
print(lst)
str1 = ' '.join(lst)
print(str1)
```

```python
# isalnum - True if all chars are alphanumeric
# isalpha - True if all chars are alphabets
# isdigits - True if all chars are numbers
# isspace - True if all chars are space
# islower, isupper, istitle 
print('123'.isdigit()) # True
print('hello'.isalpha()) # True
print('hello123'.isalnum()) # True
print('hello world 123'.islower()) # True
```

```python
# looping over strings 
str1 = 'python script_01.1'
for char in str1:
    if char.isalpha() or char.isalnum() or char.isdigit():
        res = char.upper()
        print(res,end=' ')
```

```python
# count vowels in word / string 
def count_vowels(word: str) -> float:
    counter = 0
    idx = 0
    while idx < len(word):
        if word[idx].lower() in 'aeiou':
            counter += 1
        idx += 1
    return counter
print(count_vowels('hello'))
```

```python
# frequency of each word of all alphabets & numbers
def char_freq(word: str):
    freq = {}
    for char in word:
        if char.isalnum():
            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1
    return freq
print(char_freq("current python version is 3.13"))
        
```
